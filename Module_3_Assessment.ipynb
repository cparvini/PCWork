# Traffic Flux Portion of the Assessment

u_max, u_star, rho_max, rho_star, A, B = sympy.symbols('u_max u_star rho_max rho_star A B')

eq1 = sympy.Eq( 0, u_max*rho_max*(1 - A*rho_max-B*rho_max**2) )
eq2 = sympy.Eq( 0, u_max*(1 - 2*A*rho_star-3*B*rho_star**2) )
eq3 = sympy.Eq( u_star, u_max*(1 - A*rho_star - B*rho_star**2) )

eq4 = sympy.Eq(eq2.lhs - 3*eq3.lhs, eq2.rhs - 3*eq3.rhs)

rho_sol = sympy.solve(eq4,rho_star)[0]
rho_sol2 = sympy.solve(eq2,rho_star)[0]
B_sol = sympy.solve(eq1,B)[0]
quadA = eq2.subs([(rho_star,rho_sol), (B,B_sol)])

A_sol = sympy.solve(quadA,A)
aval = A_sol[0].evalf(subs={u_star: 1.5, u_max: 2.0, rho_max: 15.})
bval = B_sol.evalf(subs={rho_max:15., A:aval})
rhoval = rho_sol2.evalf(subs={u_max:2.0, u_star:1.5, A:aval, B:bval})
print(aval,bval,rhoval)

```python
def ICs():

    ICL = np.zeros((3,1))
    ICR = np.zeros_like(ICL)
    
    ICL[0] = 1. 
    ICL[1] = 0.
    ICL[2] = 100.*1000.
    
    ICR[0] = .125
    ICR[1] = 0.
    ICR[2] = 10.*1000.
    
    return (ICL,ICR)
```

```python
def build_u(ICL,ICR):
    
    u = np.zeros((3,nx))
    
    u[0,:nx/2] = ICL[0]
    u[1,:nx/2] = ICL[1]*ICL[0]
    u[2,:nx/2] = 0.5*ICL[1]**2/ICL[0]+ICL[2]/(gamma-1)
    
    u[0,nx/2:] = ICR[0]
    u[1,nx/2:] = ICR[1]*ICR[0]
    u[2,nx/2:] = 0.5*ICR[1]**2/ICR[0]+ICR[2]/(gamma-1)
    
    uf = np.zeros((3,nx,nt))
    uf[:,:,0] = u[:,:]
    
    return uf
```

```python
def compute_f(u):
    
    u1, u2, u3 = u[0], u[1], u[2]
    f = np.array([u2,\
                  u2**2/u1+(gamma-1)*(u3-u2**2/(2*u1)),\
                  u2/u1*(u3+(gamma-1)*(u3-u2**2/(2*u1)))])
    return f
```

```python
def richtmyer(u_out, dx, dt, nx, nt, damp):
    un = u_out[:,:,0]
    un_plus = np.zeros((3,nx))
    un_minus = np.zeros((3,nx))
    
    for i in range(nt):
        un_plus[:,:-1] = .5*(un[:,1:]+un[:,:-1]) -\
        dt/(2*dx)*(compute_f(un[:,1:]) - compute_f(un[:,:-1]))
        
        un_minus[:,1:] = un_plus[:,:-1]
        
        un[:,1:-1] = un[:,1:-1] - dt/dx *\
        (compute_f(un_plus[:,1:-1]) - compute_f(un_minus[:,1:-1])) +\
        damp * (un[:,2:] - 2*un[:,1:-1] + un[:,:-2])
        
        un[:,0]  = un[:,1]
        un[:,-1] = un[:,-2]
        
        u_out[:,:,i] = un[:,:]
                
    return u_out
```

```python
def animate(frame):
    x = np.arange(0,nx)
    x = (x-nx/2)*dx
    y = varview[:,frame]
    line.set_data(x,y)
    return line,
```

```python
def get_vars(u):
    
    vel = u[1,:,:]/u[0,:,:]
    pres = (gamma - 1)*(u[2,:,:] - .5 * u[1,:,:]**2 / u[0,:,:])
    rho = u[0,:,:]
    
    return vel, pres, rho
```

```python
nx = 81
nt = 141
dx = .25
dt = .0002
gamma = 1.4

ICL,ICR = ICs()
uf = build_u(ICL,ICR)

ut = richtmyer(uf,dx,dt,nx,nt,damp=0.0)
vel,pres,rho = get_vars(ut)
print(vel[int(2.5/dx+nx/2),int(0.01/dt)-1],\
      pres[int(2.5/dx+nx/2),int(0.01/dt)-1],\
      rho[int(2.5/dx+nx/2),int(0.01/dt)-1])

varview = vel.copy()
fig = plt.figure();
ax = plt.axes(xlim=(-nx/2*dx,nx/2*dx),ylim=(0,np.amax(varview)*1.5));
line, = ax.plot([],[],lw=2);
anim = animation.FuncAnimation(fig, animate, nt, interval=30)
display_animation(anim, default_mode='loop')
```
