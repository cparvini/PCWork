```python
def get_vars(u):
    
    vel = u[1,:,:]/u[0,:,:]
    pres = (gamma - 1)*(u[2,:,:] - .5 * u[1,:,:]**2 / u[0,:,:])
    rho = u[0,:,:]
    
    return vel, pres, rho

def build_u(ICL,ICR,nx):
    
    u = numpy.zeros((3,nx))
    
    u[0,:nx/2] = ICL[0]
    u[1,:nx/2] = ICL[1]*ICL[0]
    u[2,:nx/2] = 0.5*ICL[1]**2/ICL[0]+ICL[2]/(gamma-1)
    
    u[0,nx/2:] = ICR[0]
    u[1,nx/2:] = ICR[1]*ICR[0]
    u[2,nx/2:] = 0.5*ICR[1]**2/ICR[0]+ICR[2]/(gamma-1)
    
    return u

def compute_F(u):
    
    u1, u2, u3 = u[0], u[1], u[2]
    f = numpy.array([u2,\
                  u2**2/u1+(gamma-1)*(u3-u2**2/(2*u1)),\
                  u2/u1*(u3+(gamma-1)*(u3-u2**2/(2*u1)))])
    return f

def ICs():

    ICL = numpy.zeros((3,1))
    ICR = numpy.zeros_like(ICL)
    
    ICL[0] = 1. 
    ICL[1] = 0.
    ICL[2] = 100.
    
    ICR[0] = .125
    ICR[1] = 0.
    ICR[2] = 10.
    
    return (ICL,ICR)
```

```python
def rho_green_light(nx, rho_max, rho_light):
    """Computes "green light" initial condition with shock, and linear distribution behind

    Parameters
    ----------
    nx        : int
        Number of grid points in x
    rho_max   : float
        Maximum allowed car density
    rho_light : float
        Density of cars at stoplight

    Returns
    -------
    rho: array of floats
        Array with initial values of density
    """    
    rho = numpy.arange(nx)*2./nx*rho_light  # Before stoplight
    rho[(nx-1)/2:] = 0
    
    return rho
```

```python
def godunov(rho, nt, dt, dx, rho_max, V_max):
    """ Computes the solution with the Godunov scheme using the Lax-Friedrichs flux.
    
    Parameters
    ----------
    rho    : array of floats
            Density at current time-step
    nt     : int
            Number of time steps
    dt     : float
            Time-step size
    dx     : float
            Mesh spacing
    rho_max: float
            Maximum allowed car density
    V_max  : float
            Speed limit
    
    Returns
    -------
    rho_n : array of floats
            Density after nt time steps at every point x
    """
    
    #initialize our results array with dimensions nt by nx
    rho_n = numpy.zeros((nt,len(rho)))      
    #copy the initial u array into each row of our new array
    rho_n[:,:] = rho.copy()              
    
    #setup some temporary arrays
    rho_plus = numpy.zeros_like(rho)
    rho_minus = numpy.zeros_like(rho)
    flux = numpy.zeros_like(rho)
  
    for t in range(1,nt):
        
        rho_plus[:-1] = rho[1:] # Can't do i+1/2 indices, so cell boundary
        rho_minus[:] = rho[:] # arrays at index i are at location i+1/2
        flux = 0.5 * (computeF(V_max, rho_max, rho_minus) + 
                      computeF(V_max, rho_max, rho_plus) + 
                      dx / dt * (rho_minus - rho_plus))
        rho_n[t,1:-1] = rho[1:-1] + dt/dx*(flux[:-2]-flux[1:-1])
        rho_n[t,0] = rho[0]
        rho_n[t,-1] = rho[-1]
        rho = rho_n[t].copy()
        
    return rho_n
```

```python
def minmod(e, dx, nx):
    """
    Compute the minmod approximation to the slope
    
    Parameters
    ----------
    e  : array of float 
        input data
    dx : float 
        spacestep
    nx : int
        number of spacial steps
    
    Returns
    -------
    sigma : array of float 
            minmod slope
    """
    
    sigma = numpy.zeros_like(e)
    de_minus = numpy.ones_like(e)
    de_plus = numpy.ones_like(e)
    
    de_minus[:,1:] = e[:,1:] - e[:,:-1]
    de_plus[:,:-1] = e[:,:-1] - e[:,1:]
    # The following is inefficient but easy to read
    for j in range(1,3):
        for i in range(1, nx-1):
            if (de_minus[j,i] * de_plus[j,i] < 0.0):
                sigma[j,i] = 0.0
            elif (numpy.abs(de_minus[j,i]) < numpy.abs(de_plus[j,i])):
                sigma[j,i] = de_minus[j,i]
            else:
                sigma[j,i] = de_plus[j,i]
    return sigma
```

```python
def muscl(uf, nt, dt, dx, nx):
    """ Computes the solution with the MUSCL scheme using the Lax-Friedrichs flux,
    RK2 in time and minmod slope limiting.
    
    Parameters
    ----------
    uf     : array of floats
            Array u at current time-step
    nt     : int
            Number of time steps
    dt     : float
            Time-step size
    dx     : float
            Mesh spacing
    nx     : int
            Number of spacial steps
    
    Returns
    -------
    un : array of floats
            Array u after nt time steps at every point x
    """
    
    #initialize our results array with dimensions 3 by nx by nt
    un = numpy.zeros((3,nx,nt))      
    #copy the initial u array into each row of our new array
    un[:,:,0] = uf.copy()              
    
    #setup some temporary arrays
    un_plus = numpy.zeros_like(uf)
    un_minus = numpy.zeros_like(uf)
    flux = numpy.zeros_like(uf)
    un_star = numpy.zeros_like(uf)

    for t in range(1,nt):
               
        sigma = minmod(uf,dx,nx) #calculate minmod slope

        #reconstruct values at cell boundaries
        un_plus[:,:-1] = uf[:,1:] + sigma[:,1:]*dx/2.
        un_minus = uf - sigma*dx/2.
        
        flux = 0.5 * (compute_F(un_minus) + 
                      compute_F(un_plus) + 
                      dx / dt * (un_minus - un_plus))
        
        un_star[:,1:-1] = uf[:,1:-1] + dt/dx*(flux[:,:-2]-flux[:,1:-1])
        un_star[:,0] = uf[:,0]
        un_star[:,-1] = uf[:,-1]
        
        
        sigma = minmod(un_star,dx,nx) #calculate minmod slope
    
        #reconstruct values at cell boundaries
        un_plus[:,:-1] = un_star[:,1:] + sigma[:,1:]*dx/2.
        un_minus = un_star - sigma*dx/2.
        
        
        flux = 0.5 * (compute_F(un_minus) + 
                      compute_F(un_plus) + 
                      dx / dt * (un_minus - un_plus))
        
        un[:,1:-1,t] = 0.5 * (uf[:,1:-1] + un_star[:,1:-1] + 
                               dt/dx*(flux[:,:-2]-flux[:,1:-1]))
        un[:,0,t] = uf[:,0]
        un[:,-1,t] = uf[:,-1]
        uf = un[:,:,t].copy()
        
    return un
```
